```python
import pandas as pd
import os
data = pd.read_csv(os.path.join('data','data.csv'))
data
```

```python
dict(data['instr'].value_counts())
```

```python
dict(data['class'].value_counts())
```

```python
def combine(a, b):
    return str(a) + '-' + str(b)

data['instr-class'] = list(map(lambda a, b: combine(a, b), data['instr'], data['class']))
```
